```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as se
```

```python
data=pd.read_csv("titanic.csv")
```

```python
from seaborn import load_dataset
```

```python
tips=load_dataset("tips")
tips
```

```python
data
```

```python
se.countplot(x='Survived',data=data)
plt.show()
```

```python
data["Sex"].value_counts().plot(kind='pie',autopct='%.2f')
plt.show()
```

```python
plt.hist(data['Age'],bins=5)
plt.show()
```

```python
se.distplot(data['Age'])
plt.show()
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
plt.show()
```

```python
se.scatterplot(data=tips,x='total_bill',y='tip',hue=tips['sex'])
plt.show()
```

```python
se.scatterplot(data=tips,x='total_bill',y='tip',hue=tips['sex'],style=tips['smoker'])
plt.show()
```

```python
se.barplot(x='Pclass',y='Age',data=data)
plt.show()
```

```python
se.barplot(x='Pclass',y='Age',hue='Sex',data=data)
plt.show()
```

```python
se.boxplot(x='Pclass',y='Age',data=data)
plt.show()
```

```python
se.boxplot(x='Pclass',y='Age',hue=data['Sex'],data=data)
plt.show()
```
